```python
from bs4 import BeautifulSoup
import requests
import shutil
url = "https://www.pexels.com/search/hd%20wallpaper/"
page = requests.get(url).text
soup = BeautifulSoup(page,"html.parser")
lst = []
for i in soup.findAll('a'):
    if(i.get('href')):
        lst.append(i.get('href'))
print(lst)
```

```python
    
def download_image(image, i):
    response = requests.get(image)
    realname = ''.join(e for e in image[1] if e.isalnum())
    file = open("C://Data//{}.jpg".format(realname + str(i) ), 'wb')
    response.raw.decode_content = True
    shutil.copyfileobj(response.raw, file)
    del response
```

```python
for i in range(0, len(lst)):
    download_image(lst[i],i)
```

```python
from bs4 import BeautifulSoup
import requests
import urllib.request
import shutil

url = "http://www.nitm.ac.in/gallery/2020/7th-convocation-held-on-4th-december-2020-1";
response = requests.get(url).text
soup = BeautifulSoup(response, "html.parser")
link = soup.find_all("img",class_="img-responsive")
image_info = []
for a in link:
    image_tag = a.findChildren("img")
    print(image_tag)
```

```python
def downloadimage(image):
    response = requests.get(image[0],stream=True)
    realname = ''.join(e for e in image[1] if e.isalnum())
    file = open("C://Data//{}.jpg".format(realname), 'wb')
    response.raw.decode_content = True
    shutil.copyfileobj(response.raw, file)
    del response
```

```python
for i in range(0, len(image_info)):
    downloadimage(image_info[i])
```
